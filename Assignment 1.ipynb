```python
import datetime
from tabulate import tabulate
TODO = []
```

```python
class todo:
    """A todo class"""
    def __init__(self, desc, date, place):
        self.desc = desc
        self.date = date
        self.place = place
```

```python
def Todo_Menu():
    """Displays the menu of Todo"""

    print("--------------------------------------------------------------------.........\n")
    print("Please entered the code associated with an operation to perform that operatio\n")
    print("1  ->  ADD new todo item")
    print("2  ->  Update existing todo item")
    print("3  ->  Delete a todo item")
    print("4  ->  Display entire todos")
    print("5  ->  Check all todos for today")
    print("6  ->  Store todos in a txt file\n")
    print("--------------------------------------------------------------------.........\n")
```

```python
def add():
    """Add a item to todo"""
    
    try:
        desc = input("Enter description: ")

        date_time_str = input('Enter a date and in YYYY-MM-DD HH-MM format')
        date1 = datetime.datetime.strptime(date_time_str, '%Y-%m-%d %H:%M')

        place = input("Enter place: ")

        td = todo(desc, date1, place)
    except:
        print("Invalid input")
    try:
        TODO.append(td)
        print("Todo Added Successfully")
    except:
        print("Could not add")
```

```python
def display():
    """Display list of todos"""
    """Displays the entire phonebook content"""
    cnt = 1
    li = []
    for td in TODO:
        li.append([cnt, td.desc, td.date, td.place])
        cnt += 1
    tb = tabulate(li, headers=['S.No','Description','Date','Place'])
    print(tb)
```

```python
def Check_Conflict(td):
    """Checks if there is a conflict in adding or updating a todo"""
    cnt = 0
    for x in TODO:
        if x.date == td.date:
            return True
    return False
```

```python
def Display_Todo(td):
    cnt = 1
    li = []
    li.append([cnt, td.desc, td.date, td.place])
    tb = tabulate(li, headers=['S.No','Description','Date','Place'])
    print(tb)
```

```python
def update():    
    """Updates a todo"""
    
    try:
        date_time_str = input('Enter a date and in YYYY-MM-DD HH-MM format')
        date1 = datetime.datetime.strptime(date_time_str, '%Y-%m-%d %H:%M')

        found = False

        for x in TODO:
            if x.date == date1:
                found = True
                Display_Todo(x)
                desc = input("Enter new description: ")

                date_time_str = input('Enter a date and in YYYY-MM-DD HH-MM format')
                date1 = datetime.datetime.strptime(date_time_str, '%Y-%m-%d %H:%M')

                place = input("Enter new place: ")

                td = todo(desc, date1, place) 

                if Check_Conflict(td) == True:
                    choice = input("Do you want to delete conflicted todo or update new todo information?(Y/N): ")
                    if choice == "Y":
                        print("First update or delete the previously added todo and then enter new todo")
                        break
                    else:

                        while True:
                            date_time_str = input('Enter a date and in YYYY-MM-DD HH-MM format')

                            if date_time_str == "0":
                                break
                            date1 = datetime.datetime.strptime(date_time_str, '%Y-%m-%d %H:%M')

                            place = input("Enter new place: ")

                            td = todo(desc, date1, place) 

                            if Check_Conflict(td) == False:
                                x.desc = desc
                                x.date = date1
                                x.place = place
                                print("Updated Successfully")
                                break
                else:
                    x.desc = desc
                    x.date = date1
                    x.place = place
                    print("Updated Successfully")
                    break
        if found == False:
            print("Could not find todo")
    except:
        print("There was an error updating")
```

```python
def delete():
    """Delete a todo"""
    
    try:
        date_time_str = input('Enter a date and in YYYY-MM-DD HH-MM format')
        date1 = datetime.datetime.strptime(date_time_str, '%Y-%m-%d %H:%M')
        found = False

        for i in range(len(TODO)):
            if TODO[i].date == date1:
                found = True
                del TODO[i]
                print("Successfully Deleted")
        if found == False:
            print("Todo not found")
    except:
        print("Error while deleting")
```

```python
from datetime import date
def check():
    """Checks all todos for present day"""
    try:
        today = date.today()

        li = []
        cnt = 1

        for td in TODO:
            if td.date.date() == today:
                li.append([cnt, td.desc, td.date, td.place])
                cnt += 1
        if len(li) == 0:
            print("No todos for today")
        else:
            tb = tabulate(li, headers=['S.No','Description','Date','Place'])
            print(tb)
    except:
        print("Error while searching")
```

```python
def store():
    """Stores all todos in a txt file"""
    
    try:
        cnt = 1
        li = []
        for td in TODO:
            li.append([cnt, td.desc, td.date, td.place])
            cnt += 1
        tb = tabulate(li, headers=['S.No','Description','Date','Place'])

        fname = 'Todo.txt'
        with open(fname, 'w') as file_obj:
            file_obj.write(tb)
    except:
        print("Error occured")
```

```python
def operate_todo(x):
    """Does phonebook operation"""
    
    if x == "1":
        add()
    elif x == "2":
        update()
    elif x == "3":
        delete()
    elif x == "4":
        display()
    elif x == "5":
        check()
    elif x == "6":
        store()
```

```python
while True:
    Todo_Menu()
    i = input("Enter any of the above code (1-6) to perform required operations or type quit to exit ")
    if i == "quit":
        break
    operate_todo(i)
```
