```python
Dict = []
```

```python
class word:
    def __init__(self, code, name, desc):
        self.code = code
        self.name = name
        self.desc = desc
```

```python
def Dictionary_Menu():
    """Displays the menu of Dictionary"""
    
    print("--------------------------------------------------------------------.........\n")
    print("Please entered the code associated with an operation to perform that operatio\n")
    print("1  ->  ADD new word")
    print("2  ->  Update existing word")
    print("3  ->  Delete a word")
    print("4  ->  Display entire dictionary")
    print("5  ->  Search a word")
    print("6  ->  Store words in a txt file\n")
    print("--------------------------------------------------------------------.........\n")
```

```python
def add():
    """Add a word to dictionary"""
    
    try:
        code = len(Dict)+1

        name = input("Enter name: ")
        desc = input("Enter description: ")

        w = word(code, name, desc)

        Dict.append(w)
        print("Successfully added")
        
    except:
        print("Error while adding")
```

```python
def search():
    """Search for a word in dictionary"""
    
    w = input("Enter a word to search for: ")
    
    li = []
    
    found = False
    for word in Dict:
        if word.name == w:
            found = True
            li.append([word.code, word.name, word.desc])
    
    if found == True:
        tb = tabulate(li, headers=['Code No','Name','Description'])
        print(tb)
    else:
        print("Word not found")
```

```python
def update():
    """Update a word in dictionary"""
    w = input("Enter a word to search for: ")
    
    found = False
    
    for word in Dict:
        if word.name == w:
            name = input("Enter new word name: ")
            desc = input("Enter new word description: ")
            found = True
            word.name = name
            word.desc = desc
            print("Successfully Updated")
            break
            
    if found == False:
        print("Word not found")
```

```python
def delete():
    """Delete a word from dictionary"""
    w = input("Enter a word to search for: ")
    
    try:
        found = False

        for i in range(len(Dict)):
            if Dict[i].name == w:
                found = True
                print("Succesfully Deleted")
                del Dict[i]

        if found == False:
            print("Word not found")
    except:
        print("Error while deleting")
```

```python
from tabulate import tabulate
def display():
    """Displays the entire phonebook content"""

    li = []
    for word in Dict:
        li.append([word.code, word.name, word.desc])
    tb = tabulate(li, headers=['Code No','Name','Description'])
    print(tb)
```

```python
def store():
    """Stores the dictionary into a txt file"""
    try:
        li = []
        for word in Dict:
            li.append([word.code, word.name, word.desc])
        tb = tabulate(li, headers=['Code No','Name','Description'])

        fname = 'Dictionary.txt'
        with open(fname, 'w') as file_obj:
            file_obj.write(tb)
    except:
        print("Error occured")
```

```python
def operate_dict(x):
    """Does Dictionary operation"""
    
    if x == "1":
        add()
    elif x == "2":
        update()
    elif x == "3":
        delete()
    elif x == "4":
        display()
    elif x == "5":
        search()
    elif x == "6":
        store()
```

```python
while True:
    Dictionary_Menu()
    i = input("Enter any of the above code (1-6) to perform required operations or type quit to exit ")
    if i == "quit":
        break
    operate_dict(i)
```
